# visualizing repositories using pygal
 this is an introductory notebook on how to visualize the data  generated by github api  for analyzing the data. here we will find the relative popularity of python projects(or any other language) present on github

```python
#requests is used for fetching data using http protocol and pygal is
#data visualizing library
import requests
import pygal 
from pygal.style import LightColorizedStyle as LCS ,LightenStyle as LS


```

```python
# fetching the data and checking the status code
r = requests.get('https://api.github.com/search/repositories?q=language:python&sort=star')
print("status code:",r.status_code)
```

```python
#storing an api in a variable in json format and displaying the information
response_dicts = r.json()
print("total repo ",response_dicts['total_count'])
repo_dicts = response_dicts['items']
names , stars = [] ,[]
for repo_info in repo_dicts:
    names.append(repo_info['name'])
    stars.append(repo_info['stargazers_count'])

#making corresponding visualization
my_style = LS('#333366',base_style=LCS)
chart = pygal.Bar(style=my_style, x_label_rotation=45, show_legend=False)
chart.title = 'trending projects in python on github'
chart.add('',stars)
chart.render_to_file('python_repo.svg')
```
